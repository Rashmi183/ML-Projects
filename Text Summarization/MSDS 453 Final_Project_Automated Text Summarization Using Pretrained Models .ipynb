#### Using facebook/bart-base Model

### Details
The facebook/bart-base model is a pre-trained sequence-to-sequence transformer developed by Facebook AI. It features a bidirectional encoder, similar to BERT, and an autoregressive decoder, akin to GPT. The model is trained to reconstruct original text from corrupted versions, making it particularly effective for tasks like text generation, summarization, translation, text classification, and question answering. Models range from BART-base (6 encoder/decoder layers).Combination of BERT-like bidirectional encoder and GPT-like autoregressive decoder.Encoder processes input bidirectionally (like BERT), and the decoder generates text autoregressively (like GPT).Ideal for tasks requiring both understanding and generation of text.

#### Training Hyperparameters
BATCH_SIZE = 8
The number of samples processed together in a single forward and backward pass. A small batch size (like 8) is often used due to GPU memory limitations.

EPOCHS = 5
The number of times the entire training dataset is passed through the model. Increasing the epochs may improve accuracy but can also lead to overfitting if too large.

MAX_LENGTH = 64
The maximum sequence length of input tokens. Texts longer than this limit will be truncated, and shorter ones will be padded. This ensures uniformity in batch processing.

GRADIENT_ACCUMULATION_STEPS = 32
The number of steps to accumulate gradients before performing a backward pass. This effectively increases the batch size (logical batch size = BATCH_SIZE * GRADIENT_ACCUMULATION_STEPS) without requiring additional memory.

#### Optimization Settings
NUM_WORKERS = 0
The number of CPU workers for loading data. 0 means data loading happens on the main process/thread. Increasing this can speed up data loading for larger datasets.

EVAL_STEPS = 500
The frequency (in training steps) at which the model is evaluated on the validation set. This helps monitor progress and performance during training.

MAX_GRAD_NORM = 0.5
The maximum norm of gradients to prevent exploding gradients (a common problem in training deep neural networks). Gradient clipping ensures stability during optimization.

LEARNING_RATE = 3e-5
The step size for the optimizer to adjust the model weights during training. A smaller value ensures gradual updates and helps avoid overshooting the optimal weights.

BF16 stands for Brain Floating Point 16. It's a 16-bit floating-point format used in machine learning to reduce memory usage and accelerate computations, particularly on hardware accelerators like GPUs and TPUs.BF16 is commonly used to train large neural networks, especially on hardware accelerators like GPUs and TPUs.

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

import torch
import pandas as pd
import numpy as np
import logging
from tqdm import tqdm
import matplotlib.pyplot as plt
from transformers import (
    BartTokenizer,
    BartForConditionalGeneration,
    TrainingArguments,
    Trainer
)
from rouge_score import rouge_scorer
import spacy
from collections import Counter

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

class Config:
    MODEL = 'facebook/bart-base'
    OUT_DIR = "./bart_checkpoints"
    BATCH_SIZE = 8
    EPOCHS = 5
    MAX_LENGTH = 64
    GRADIENT_ACCUMULATION_STEPS = 32
    NUM_WORKERS = 0
    EVAL_STEPS = 500
    MAX_GRAD_NORM = 0.5
    LEARNING_RATE = 3e-5

class MemoryEfficientMeetingDataset(torch.utils.data.Dataset):
    def __init__(self, df, tokenizer, max_length):
        self.df = df
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        text = str(self.df.iloc[idx, 0])[:self.max_length * 4]
        summary = str(self.df.iloc[idx, 1])[:self.max_length * 2]

        input_encoding = self.tokenizer(
            text,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        target_encoding = self.tokenizer(
            summary,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        return {
            'input_ids': input_encoding['input_ids'].squeeze(),
            'attention_mask': input_encoding['attention_mask'].squeeze(),
            'labels': target_encoding['input_ids'].squeeze()
        }

class CustomTrainer(Trainer):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.training_history = []

    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        loss = super().compute_loss(model, inputs, return_outputs)

        if self.state.global_step % self.args.logging_steps == 0:
            self.training_history.append({
                'step': self.state.global_step,
                'training_loss': loss.item()
            })

        return loss

    def evaluation_loop(self, *args, **kwargs):
        output = super().evaluation_loop(*args, **kwargs)

        current_train_loss = self.training_history[-1]['training_loss'] if self.training_history else 0

        if self.state.global_step % self.args.eval_steps == 0:
            metrics = {
                'step': self.state.global_step,
                'training_loss': current_train_loss,
                'validation_loss': output.metrics['eval_loss']
            }

        return output

def clean_memory():
    """Clean up GPU memory."""
    torch.cuda.empty_cache()

def load_data(file_path, max_samples=None):
    """Load and preprocess the dataset."""
    logger.info(f"Loading data from {file_path}...")
    df = pd.read_csv(
        file_path,
        encoding='ISO-8859-1',
        on_bad_lines='skip',
        nrows=max_samples,
        low_memory=False
    )
    df.drop(['meeting_id', 'type', 'city'], axis=1, errors='ignore', inplace=True)
    df = df.apply(lambda x: x.str.strip() if isinstance(x, str) else x)
    df = df.fillna('')
    logger.info(f"Loaded {len(df)} rows after cleaning.")
    return df

def generate_summary(text, model, tokenizer, max_length=64):
    """Generate a summary."""
    with torch.no_grad():
        inputs = tokenizer.encode(
            text[:max_length * 4],
            return_tensors='pt',
            max_length=max_length * 2,
            truncation=True
        )

        if torch.cuda.is_available():
            inputs = inputs.cuda()

        summary_ids = model.generate(
            inputs,
            max_length=max_length,
            num_beams=4,
            length_penalty=2.0,
            early_stopping=True,
            no_repeat_ngram_size=3
        )

        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary

def calculate_rouge_scores(predicted_summary, reference_summary):
    """Calculate ROUGE scores for a pair of summaries."""
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference_summary, predicted_summary)
    return {
        'rouge1': scores['rouge1'].fmeasure,
        'rouge2': scores['rouge2'].fmeasure,
        'rougeL': scores['rougeL'].fmeasure
    }

def plot_rouge_scores(avg_rouge_scores, width=50, height=10):
    """Display average ROUGE scores as a console-based bar chart."""
    print("\nAverage ROUGE Scores Visualization")
    print("=" * 80)

    metrics = list(avg_rouge_scores.keys())
    scores = list(avg_rouge_scores.values())
    max_score = max(scores)

    y_axis = [f"{i/10:.1f}" for i in range(0, 11)]
    max_y_label_length = max(len(label) for label in y_axis)

    for i in range(height - 1, -1, -1):
        current_value = i / (height - 1)
        print(f"{f'{current_value:.1f}'.rjust(max_y_label_length)} │", end=" ")

        for score in scores:
            if score >= current_value:
                print("█", end="  ")
            else:
                print(" ", end="  ")
        print()

    print(" " * max_y_label_length + "└" + "─" * (len(metrics) * 3))
    print(" " * (max_y_label_length + 1) + "  ".join(metrics))

    print("\nActual Values:")
    for metric, score in avg_rouge_scores.items():
        print(f"{metric}: {score:.3f}")

def display_console_histogram(rouge_scores, width=50):
    """Display a histogram of ROUGE scores in the console."""
    print("\nROUGE Scores Distribution")
    print("=" * 80)

    for metric, scores in rouge_scores.items():
        scores_array = np.array(scores) * 100

        mean_val = np.mean(scores_array)
        median_val = np.median(scores_array)
        std_val = np.std(scores_array)

        print(f"\n{metric.upper()}:")
        print("-" * 70)
        print(f"Mean: {mean_val:.2f}% | Median: {median_val:.2f}% | Std Dev: {std_val:.2f}%")

        hist, bin_edges = np.histogram(scores_array, bins=10)
        max_count = max(hist)

        for count, edge in zip(hist, bin_edges[:-1]):
            bar_length = int((count / max_count) * width)
            bar = "█" * bar_length + "░" * (width - bar_length)
            next_edge = bin_edges[np.where(bin_edges == edge)[0][0] + 1]
            print(f"{edge:5.1f}%-{next_edge:5.1f}% |{bar}| ({count:3d})")

        print("-" * 70)

def display_summary_example(original_text, generated_summary, reference_summary, rouge_scores, idx):
    """Display a formatted example of summarization."""
    print("\n" + "=" * 100)
    print(f"EXAMPLE SUMMARY #{idx+1}")
    print("=" * 100)

    print("\n📝 ORIGINAL TEXT:")
    print("-" * 40)
    print(original_text[:500] + "..." if len(original_text) > 500 else original_text)

    print("\n📋 GENERATED SUMMARY:")
    print("-" * 40)
    print(generated_summary)

    print("\n📌 REFERENCE SUMMARY:")
    print("-" * 40)
    print(reference_summary)

    print("\n📊 ROUGE SCORES:")
    print("-" * 40)
    for metric, score in rouge_scores.items():
        percentage = score * 100
        bar_length = int(percentage / 2)
        bar = "█" * bar_length + "░" * (50 - bar_length)
        print(f"{metric:<10} {percentage:>6.2f}% |{bar}|")

    print("\n" + "=" * 100)

def create_ascii_wordcloud(texts, width=80):
    """Generate and display ASCII word cloud in console."""
    doc = nlp(' '.join(texts))
    words = [token.lemma_.lower() for token in doc
             if not token.is_stop and not token.is_punct
             and token.text.strip() and len(token.text) > 2]

    word_freq = Counter(words)
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    max_freq = max(freq for _, freq in sorted_words)

    print("\nWord Frequency Analysis")
    print("=" * width)
    print(f"{'Word':<20} {'Frequency':<10} {'Distribution':<50}")
    print("-" * width)

    for word, freq in sorted_words[:20]:
        scaled_freq = int((freq / max_freq) * 40)
        bar = "█" * scaled_freq
        percentage = (freq / sum(freq for _, freq in sorted_words)) * 100
        print(f"{word:<20} {freq:<10} {bar:<40} {percentage:>6.2f}%")

    print("=" * width)
    print(f"Total unique words analyzed: {len(word_freq)}")
    print(f"Total words processed: {sum(word_freq.values())}")

def train_model(max_train_samples=1000, max_val_samples=100):
    """Train the BART model."""
    try:
        clean_memory()

        train_df = load_data("/content/sample_data/train.csv", max_samples=max_train_samples)
        val_df = load_data("/content/sample_data/val.csv", max_samples=max_val_samples)

        tokenizer = BartTokenizer.from_pretrained(Config.MODEL)
        model = BartForConditionalGeneration.from_pretrained(
            Config.MODEL,
            torch_dtype=torch.bfloat16,
            low_cpu_mem_usage=True
        )

        if torch.cuda.is_available():
            model = model.cuda()

        train_dataset = MemoryEfficientMeetingDataset(train_df, tokenizer, Config.MAX_LENGTH)
        val_dataset = MemoryEfficientMeetingDataset(val_df, tokenizer, Config.MAX_LENGTH)

        training_args = TrainingArguments(
            output_dir=Config.OUT_DIR,
            num_train_epochs=Config.EPOCHS,
            per_device_train_batch_size=Config.BATCH_SIZE,
            per_device_eval_batch_size=Config.BATCH_SIZE,
            warmup_steps=100,
            weight_decay=0.01,
            logging_dir="./logs",
            logging_steps=100,
            evaluation_strategy='steps',
            eval_steps=200,
            save_strategy='steps',
            save_steps=200,
            save_total_limit=1,
            learning_rate=Config.LEARNING_RATE,
            dataloader_num_workers=Config.NUM_WORKERS,
            bf16=True,
            gradient_accumulation_steps=Config.GRADIENT_ACCUMULATION_STEPS,
            load_best_model_at_end=True,
            max_grad_norm=Config.MAX_GRAD_NORM,
            report_to="none"
        )

        trainer = CustomTrainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset
        )

        trainer.train()

        return model, tokenizer, val_df

    except Exception as e:
        logger.error(f"Error during training: {str(e)}")
        raise

def analyze_results(generated_summaries, reference_summaries, rouge_scores_list):
    """Analyze and display comprehensive results."""
    print("\nSUMMARIZATION ANALYSIS")
    print("=" * 80)

    gen_lengths = [len(summary.split()) for summary in generated_summaries]
    ref_lengths = [len(summary.split()) for summary in reference_summaries]

    print("\n📏 Length Statistics:")
    print(f"Generated Summaries - Avg: {np.mean(gen_lengths):.1f} words, "
          f"Min: {min(gen_lengths)} words, Max: {max(gen_lengths)} words")
    print(f"Reference Summaries - Avg: {np.mean(ref_lengths):.1f} words, "
          f"Min: {min(ref_lengths)} words, Max: {max(ref_lengths)} words")

    print("\n📈 ROUGE Score Statistics:")
    for metric in rouge_scores_list[0].keys():
        scores = [scores[metric] for scores in rouge_scores_list]
        print(f"\n{metric}:")
        print(f"  Mean: {np.mean(scores):.4f}")
        print(f"  Median: {np.median(scores):.4f}")
        print(f"  Std Dev: {np.std(scores):.4f}")
        print(f"  95% Confidence Interval: [{np.percentile(scores, 2.5):.4f}, "
              f"{np.percentile(scores, 97.5):.4f}]")

def main():
    """Main execution pipeline."""
    try:
        logger.info("Starting training...")
        model, tokenizer, val_data = train_model(max_train_samples=1000, max_val_samples=100)

        logger.info("Generating summaries and calculating ROUGE scores...")
        combined_rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}
        generated_summaries = []
        reference_summaries = []
        rouge_scores_list = []

        chunk_size = len(val_data) // 5  # Process in chunks for memory efficiency

        for i in tqdm(range(0, len(val_data), chunk_size), desc="Processing Chunks"):
            chunk_end = min(i + chunk_size, len(val_data))

            for j in range(i, chunk_end):
                original_text = val_data.iloc[j, 0]
                reference_summary = val_data.iloc[j, 1]
                reference_summaries.append(reference_summary)

                generated_summary = generate_summary(original_text, model, tokenizer)
                generated_summaries.append(generated_summary)

                rouge_scores = calculate_rouge_scores(generated_summary, reference_summary)
                rouge_scores_list.append(rouge_scores)

                for metric in combined_rouge_scores:
                    combined_rouge_scores[metric].append(rouge_scores[metric])

                if j < 3:  # Display first 3 examples
                    display_summary_example(original_text, generated_summary,
                                         reference_summary, rouge_scores, j)

            clean_memory()  # Clean up memory after each chunk

        # Display ROUGE score distributions
        display_console_histogram(combined_rouge_scores)

        # Calculate average ROUGE scores
        avg_rouge_scores = {metric: np.mean(scores)
                          for metric, scores in combined_rouge_scores.items()}

        # Display histogram for average ROUGE scores
        plt.figure(figsize=(8, 6))
        metrics = list(avg_rouge_scores.keys())
        scores = list(avg_rouge_scores.values())
        plt.bar(metrics, scores, color=['#2ecc71', '#3498db', '#9b59b6'])
        plt.title('Average ROUGE Scores')
        plt.xlabel('ROUGE Metric')
        plt.ylabel('Average Score')
        plt.ylim(0, 1)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

        print("\nFinal ROUGE Scores:")
        for metric, score in avg_rouge_scores.items():
            print(f"{metric}: {score:.4f}")

        # Display word frequency analysis
        create_ascii_wordcloud(generated_summaries)

        # Display comprehensive analysis
        analyze_results(generated_summaries, reference_summaries, rouge_scores_list)

        logger.info("Process completed successfully!")

    except Exception as e:
        logger.error(f"An error occurred: {str(e)}")
        raise
    finally:
        clean_memory()

if __name__ == "__main__":
    main()
```

#### Using facebook/bart-base Model with RAG Retriever

### Details
Dense Passage Retrieval (DPR) model, which is widely used in information retrieval systems, especially for tasks like open-domain question answering. DPR was developed by Facebook AI and uses dense vector representations for both questions and passages, enabling efficient and accurate retrieval.

Abstractive Summarization (Indirect Use)
DPR may indirectly contribute to abstractive summarization (where summaries are generated by paraphrasing and rephrasing) in a few ways:

Retrieving Relevant Contexts:
In long documents, not all content is relevant to the summary. DPR can help:

Retrieve the most relevant passages using DPRContextEncoder based on a summarization query.
Pass these selected passages to an abstractive summarization model (e.g., BART or T5) to generate the final summary.
Scalability: Efficient similarity search (e.g., with FAISS) makes it scalable for large datasets.
By combining DPR for retrieval with models like BART or T5 for generation, we can achieve efficient and focused summarization tailored to specific queries or document sections.

The terms TOP_K_RETRIEVAL and VECTOR_DIMENSION are typically used in the context of dense vector retrieval systems, such as those found in neural-based search or retrieval pipelines.This parameter specifies the number of top results to retrieve from a vector similarity search.Refers to the size of the dense vector representation (number of dimensions) used to encode text, images, or other data.MiniLM models (like those from Sentence Transformers) often use 384, as they are optimized for smaller, faster embeddings.

```python

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

import torch
import pandas as pd
import numpy as np
import logging
from tqdm import tqdm
import matplotlib.pyplot as plt  # Added for plotting
from transformers import (
    BartTokenizer,
    BartForConditionalGeneration,
    TrainingArguments,
    Trainer,
    DPRQuestionEncoder,
    DPRContextEncoder
)
from sentence_transformers import SentenceTransformer
import faiss
from rouge_score import rouge_scorer
import spacy
from collections import Counter

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

class Config:
    MODEL = 'facebook/bart-base'
    RETRIEVER_MODEL = 'facebook/dpr-question_encoder-single-nq-base'
    CONTEXT_MODEL = 'facebook/dpr-ctx_encoder-single-nq-base'
    SENTENCE_MODEL = 'sentence-transformers/all-MiniLM-L6-v2'
    OUT_DIR = "./bart_checkpoints"
    BATCH_SIZE = 8
    EPOCHS = 5
    MAX_LENGTH = 64
    GRADIENT_ACCUMULATION_STEPS = 32
    NUM_WORKERS = 0
    EVAL_STEPS = 500
    MAX_GRAD_NORM = 0.5
    LEARNING_RATE = 3e-5
    TOP_K_RETRIEVAL = 3
    VECTOR_DIMENSION = 384

class RAGRetriever:
    def __init__(self, reference_texts):
        self.reference_texts = reference_texts
        self.encoder = SentenceTransformer(Config.SENTENCE_MODEL)
        self.index = faiss.IndexFlatL2(Config.VECTOR_DIMENSION)
        logger.info("Encoding reference texts for RAG...")
        embeddings = self.encoder.encode(reference_texts, show_progress_bar=True)
        self.index.add(np.array(embeddings).astype('float32'))

    def retrieve(self, query_text, k=Config.TOP_K_RETRIEVAL):
        query_embedding = self.encoder.encode([query_text])[0].reshape(1, -1).astype('float32')
        distances, indices = self.index.search(query_embedding, k)
        retrieved_texts = [self.reference_texts[idx] for idx in indices[0]]
        return retrieved_texts

class MemoryEfficientMeetingDataset(torch.utils.data.Dataset):
    def __init__(self, df, tokenizer, max_length, retriever=None):
        self.df = df
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.retriever = retriever

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        text = str(self.df.iloc[idx, 0])[:self.max_length * 4]
        summary = str(self.df.iloc[idx, 1])[:self.max_length * 2]

        if self.retriever is not None:
            retrieved_texts = self.retriever.retrieve(text)
            augmented_text = "\n\nSimilar examples:\n" + "\n".join(retrieved_texts) + "\n\nCurrent text:\n" + text
            text = augmented_text

        input_encoding = self.tokenizer(
            text,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        target_encoding = self.tokenizer(
            summary,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        return {
            'input_ids': input_encoding['input_ids'].squeeze(),
            'attention_mask': input_encoding['attention_mask'].squeeze(),
            'labels': target_encoding['input_ids'].squeeze()
        }

class CustomTrainer(Trainer):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.training_history = []

    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        loss = super().compute_loss(model, inputs, return_outputs)

        if self.state.global_step % self.args.logging_steps == 0:
            self.training_history.append({
                'step': self.state.global_step,
                'training_loss': loss.item()
            })

        return loss

    def evaluation_loop(self, *args, **kwargs):
        output = super().evaluation_loop(*args, **kwargs)

        current_train_loss = self.training_history[-1]['training_loss'] if self.training_history else 0

        if self.state.global_step % self.args.eval_steps == 0:
            metrics = {
                'step': self.state.global_step,
                'training_loss': current_train_loss,
                'validation_loss': output.metrics['eval_loss']
            }

        return output

def clean_memory():
    """Clean up GPU memory."""
    torch.cuda.empty_cache()

def load_data(file_path, max_samples=None):
    """Load and preprocess the dataset."""
    logger.info(f"Loading data from {file_path}...")
    df = pd.read_csv(
        file_path,
        encoding='ISO-8859-1',
        on_bad_lines='skip',
        nrows=max_samples,
        low_memory=False
    )
    df.drop(['meeting_id', 'type', 'city'], axis=1, errors='ignore', inplace=True)
    df = df.apply(lambda x: x.str.strip() if isinstance(x, str) else x)
    df = df.fillna('')
    logger.info(f"Loaded {len(df)} rows after cleaning.")
    return df

def generate_summary(text, model, tokenizer, retriever=None, max_length=64):
    """Generate a summary using RAG if retriever is available."""
    with torch.no_grad():
        if retriever is not None:
            retrieved_texts = retriever.retrieve(text)
            prompt = "Use the following similar examples to help generate a better summary.\n\n"
            prompt += "Similar examples:\n" + "\n---\n".join(retrieved_texts)
            prompt += "\n\nBased on these examples, summarize the following text:\n" + text
            text = prompt

        inputs = tokenizer.encode(
            text[:max_length * 4],
            return_tensors='pt',
            max_length=max_length * 2,
            truncation=True
        )

        if torch.cuda.is_available():
            inputs = inputs.cuda()

        summary_ids = model.generate(
            inputs,
            max_length=max_length,
            num_beams=4,
            length_penalty=2.0,
            early_stopping=True,
            no_repeat_ngram_size=3
        )

        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary

def calculate_rouge_scores(predicted_summary, reference_summary):
    """Calculate ROUGE scores for a pair of summaries."""
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference_summary, predicted_summary)
    return {
        'rouge1': scores['rouge1'].fmeasure,
        'rouge2': scores['rouge2'].fmeasure,
        'rougeL': scores['rougeL'].fmeasure
    }

def plot_rouge_scores(avg_rouge_scores, width=50, height=10):
    """Display average ROUGE scores as a console-based bar chart."""
    print("\nAverage ROUGE Scores Visualization")
    print("=" * 80)

    # Get scores and maximum value
    metrics = list(avg_rouge_scores.keys())
    scores = list(avg_rouge_scores.values())
    max_score = max(scores)

    # Create vertical axis labels
    y_axis = [f"{i/10:.1f}" for i in range(0, 11)]
    max_y_label_length = max(len(label) for label in y_axis)

    # Draw the chart from top to bottom
    for i in range(height - 1, -1, -1):
        current_value = i / (height - 1)
        # Print y-axis labels
        print(f"{f'{current_value:.1f}'.rjust(max_y_label_length)} │", end=" ")

        # Print bars
        for score in scores:
            if score >= current_value:
                print("█", end="  ")
            else:
                print(" ", end="  ")
        print()

    # Print x-axis
    print(" " * max_y_label_length + "└" + "─" * (len(metrics) * 3))
    print(" " * (max_y_label_length + 1) + "  ".join(metrics))

    # Print actual values
    print("\nActual Values:")
    for metric, score in avg_rouge_scores.items():
        print(f"{metric}: {score:.3f}")

def display_console_histogram(rouge_scores, width=50):
    """Display a histogram of ROUGE scores in the console."""
    print("\nROUGE Scores Distribution")
    print("=" * 80)

    # Create bins and count frequencies for each metric
    for metric, scores in rouge_scores.items():
        scores_array = np.array(scores) * 100  # Convert to percentages

        # Calculate statistics
        mean_val = np.mean(scores_array)
        median_val = np.median(scores_array)
        std_val = np.std(scores_array)

        print(f"\n{metric.upper()}:")
        print("-" * 70)
        print(f"Mean: {mean_val:.2f}% | Median: {median_val:.2f}% | Std Dev: {std_val:.2f}%")

        # Create histogram bins
        hist, bin_edges = np.histogram(scores_array, bins=10)
        max_count = max(hist)

        # Display histogram
        for count, edge in zip(hist, bin_edges[:-1]):
            bar_length = int((count / max_count) * width)
            bar = "█" * bar_length + "░" * (width - bar_length)
            next_edge = bin_edges[np.where(bin_edges == edge)[0][0] + 1]
            print(f"{edge:5.1f}%-{next_edge:5.1f}% |{bar}| ({count:3d})")

        print("-" * 70)

def display_avg_rouge_histogram(avg_rouge_scores, width=50):
    """Display histogram of average ROUGE scores."""
    print("\nAverage ROUGE Scores Histogram")
    print("=" * 80)

    # Convert to percentages
    scores = {k: v * 100 for k, v in avg_rouge_scores.items()}
    max_score = max(scores.values())

    # Sort metrics by score
    sorted_metrics = sorted(scores.items(), key=lambda x: x[1], reverse=True)

    print(f"\nScore Distribution (Total width: {width} blocks)")
    print("-" * 80)

    for metric, score in sorted_metrics:
        bar_length = int((score / max_score) * width)
        bar = "█" * bar_length + "░" * (width - bar_length)
        print(f"{metric:<10} [{score:>6.2f}%] |{bar}| ({bar_length}/{width})")

    print("-" * 80)
    print("Legend: █ = score value  ░ = remaining space")
    print(f"Scale: Each block ≈ {max_score/width:.2f}%")

def create_ascii_wordcloud(texts, width=80):
    """Generate and display ASCII word cloud in console."""
    doc = nlp(' '.join(texts))
    words = [token.lemma_.lower() for token in doc
             if not token.is_stop and not token.is_punct
             and token.text.strip() and len(token.text) > 2]

    word_freq = Counter(words)
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    max_freq = max(freq for _, freq in sorted_words)

    print("\nWord Frequency Analysis")
    print("=" * width)
    print(f"{'Word':<20} {'Frequency':<10} {'Distribution':<50}")
    print("-" * width)

    for word, freq in sorted_words[:20]:
        scaled_freq = int((freq / max_freq) * 40)
        bar = "█" * scaled_freq
        percentage = (freq / sum(freq for _, freq in sorted_words)) * 100
        print(f"{word:<20} {freq:<10} {bar:<40} {percentage:>6.2f}%")

    print("=" * width)
    print(f"Total unique words analyzed: {len(word_freq)}")
    print(f"Total words processed: {sum(word_freq.values())}")

def display_summary_example(original_text, retrieved_texts, generated_summary, reference_summary, rouge_scores, idx):
    """Display a formatted example of RAG-enhanced summarization."""
    print("\n" + "=" * 100)
    print(f"EXAMPLE SUMMARY #{idx+1}")
    print("=" * 100)

    print("\n📝 ORIGINAL TEXT:")
    print("-" * 40)
    print(original_text[:500] + "..." if len(original_text) > 500 else original_text)

    print("\n🔍 RETRIEVED SIMILAR TEXTS:")
    print("-" * 40)
    for i, text in enumerate(retrieved_texts, 1):
        print(f"Similar Example #{i}:")
        print(text[:200] + "..." if len(text) > 200 else text)
        print()

    print("\n📋 GENERATED SUMMARY:")
    print("-" * 40)
    print(generated_summary)

    print("\n📌 REFERENCE SUMMARY:")
    print("-" * 40)
    print(reference_summary)

    print("\n📊 ROUGE SCORES:")
    print("-" * 40)
    for metric, score in rouge_scores.items():
        percentage = score * 100
        bar_length = int(percentage / 2)
        bar = "█" * bar_length + "░" * (50 - bar_length)
        print(f"{metric:<10} {percentage:>6.2f}% |{bar}|")

    print("\n" + "=" * 100)

def train_model(max_train_samples=1000, max_val_samples=100):
    """Train the BART model with RAG enhancement."""
    try:
        clean_memory()

        train_df = load_data("/content/sample_data/train.csv", max_samples=max_train_samples)
        val_df = load_data("/content/sample_data/val.csv", max_samples=max_val_samples)

        logger.info("Initializing RAG retriever...")
        retriever = RAGRetriever(train_df.iloc[:, 0].tolist())

        tokenizer = BartTokenizer.from_pretrained(Config.MODEL)
        model = BartForConditionalGeneration.from_pretrained(
            Config.MODEL,
            torch_dtype=torch.bfloat16,
            low_cpu_mem_usage=True
        )

        if torch.cuda.is_available():
            model = model.cuda()

        train_dataset = MemoryEfficientMeetingDataset(train_df, tokenizer, Config.MAX_LENGTH, retriever)
        val_dataset = MemoryEfficientMeetingDataset(val_df, tokenizer, Config.MAX_LENGTH, retriever)

        training_args = TrainingArguments(
            output_dir=Config.OUT_DIR,
            num_train_epochs=Config.EPOCHS,
            per_device_train_batch_size=Config.BATCH_SIZE,
            per_device_eval_batch_size=Config.BATCH_SIZE,
            warmup_steps=100,
            weight_decay=0.01,
            logging_dir="./logs",
            logging_steps=100,
            evaluation_strategy='steps',
            eval_steps=200,
            save_strategy='steps',
            save_steps=200,
            save_total_limit=1,
            learning_rate=Config.LEARNING_RATE,
            dataloader_num_workers=Config.NUM_WORKERS,
            bf16=True,
            gradient_accumulation_steps=Config.GRADIENT_ACCUMULATION_STEPS,
            load_best_model_at_end=True,
            max_grad_norm=Config.MAX_GRAD_NORM,
            report_to="none"
        )

        trainer = CustomTrainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset
        )

        trainer.train()

        return model, tokenizer, val_df, retriever

    except Exception as e:
        logger.error(f"Error during training: {str(e)}")
        raise

def analyze_results(generated_summaries, reference_summaries, rouge_scores_list):
    """Analyze and display comprehensive results."""
    print("\nSUMMARIZATION ANALYSIS")
    print("=" * 80)

    gen_lengths = [len(summary.split()) for summary in generated_summaries]
    ref_lengths = [len(summary.split()) for summary in reference_summaries]

    print("\n📏 Length Statistics:")
    print(f"Generated Summaries - Avg: {np.mean(gen_lengths):.1f} words, "
          f"Min: {min(gen_lengths)} words, Max: {max(gen_lengths)} words")
    print(f"Reference Summaries - Avg: {np.mean(ref_lengths):.1f} words, "
          f"Min: {min(ref_lengths)} words, Max: {max(ref_lengths)} words")

    print("\n📈 ROUGE Score Statistics:")
    for metric in rouge_scores_list[0].keys():
        scores = [scores[metric] for scores in rouge_scores_list]
        print(f"\n{metric}:")
        print(f"  Mean: {np.mean(scores):.4f}")
        print(f"  Median: {np.median(scores):.4f}")
        print(f"  Std Dev: {np.std(scores):.4f}")
        print(f"  95% Confidence Interval: [{np.percentile(scores, 2.5):.4f}, "
              f"{np.percentile(scores, 97.5):.4f}]")
def main():
    """Main execution pipeline with RAG enhancement."""
    try:
        logger.info("Starting training with RAG enhancement...")
        model, tokenizer, val_data, retriever = train_model(max_train_samples=1000, max_val_samples=100)

        logger.info("Generating summaries and calculating ROUGE scores...")
        combined_rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}
        generated_summaries = []
        reference_summaries = []
        rouge_scores_list = []

        chunk_size = len(val_data) // 5  # Process in chunks for memory efficiency

        for i in tqdm(range(0, len(val_data), chunk_size), desc="Processing Chunks"):
            chunk_end = min(i + chunk_size, len(val_data))

            for j in range(i, chunk_end):
                original_text = val_data.iloc[j, 0]
                reference_summary = val_data.iloc[j, 1]
                reference_summaries.append(reference_summary)

                retrieved_texts = retriever.retrieve(original_text)
                generated_summary = generate_summary(original_text, model, tokenizer, retriever)
                generated_summaries.append(generated_summary)

                rouge_scores = calculate_rouge_scores(generated_summary, reference_summary)
                rouge_scores_list.append(rouge_scores)

                for metric in combined_rouge_scores:
                    combined_rouge_scores[metric].append(rouge_scores[metric])

                if j < 3:  # Display first 3 examples
                    display_summary_example(original_text, retrieved_texts, generated_summary,
                                         reference_summary, rouge_scores, j)

            clean_memory()  # Clean up memory after each chunk

        # Display ROUGE score distributions
        display_console_histogram(combined_rouge_scores)

        # Calculate average ROUGE scores
        avg_rouge_scores = {metric: np.mean(scores)
                          for metric, scores in combined_rouge_scores.items()}

        # Display histogram for average ROUGE scores
        plt.figure(figsize=(8, 6))
        metrics = list(avg_rouge_scores.keys())
        scores = list(avg_rouge_scores.values())
        plt.bar(metrics, scores, color=['#2ecc71', '#3498db', '#9b59b6'])
        plt.title('Average ROUGE Scores')
        plt.xlabel('ROUGE Metric')
        plt.ylabel('Average Score')
        plt.ylim(0, 1)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()

        print("\nFinal ROUGE Scores:")
        for metric, score in avg_rouge_scores.items():
            print(f"{metric}: {score:.4f}")

        # Display word frequency analysis
        create_ascii_wordcloud(generated_summaries)

        # Display comprehensive analysis
        analyze_results(generated_summaries, reference_summaries, rouge_scores_list)

        logger.info("Process completed successfully!")

    except Exception as e:
        logger.error(f"An error occurred: {str(e)}")
        raise
    finally:
        clean_memory()

if __name__ == "__main__":
    main()
```

#### Using t5-small Model With RAG Retriever

### Details
DPR is a retrieval-based model designed for open-domain question answering.
It consists of two encoders:
DPRQuestionEncoder: Encodes the input question into a dense vector representation.
DPRContextEncoder: Encodes passages or contexts (from a corpus) into dense vectors.
These dense vectors are used to measure similarity (via dot product) to retrieve relevant passages for the question.
Tokenizers:
DPRQuestionEncoderTokenizer: Prepares the input text for the question encoder.
DPRContextEncoderTokenizer: Prepares the input text for the context encoder.
Both tokenizers ensure that the inputs are tokenized consistently for their respective encoders.
Combining DPR with T5-small enhances performance in open-domain question-answering systems, leveraging DPR’s retrieval capability and T5’s text generation prowess.

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

import torch
import numpy as np
import pandas as pd
import evaluate
from tqdm import tqdm
import logging
import gc
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback,
    DPRQuestionEncoder,
    DPRContextEncoder,
    DPRQuestionEncoderTokenizer,
    DPRContextEncoderTokenizer
)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from rouge_score import rouge_scorer
import spacy
from typing import List, Dict, Any
import faiss
import torch.nn.functional as F

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Configuration
MODEL = 't5-small'
OUT_DIR = "./model_checkpoints"
BATCH_SIZE = 1
EPOCHS = 2
MAX_LENGTH = 64
GRADIENT_ACCUMULATION_STEPS = 32
NUM_WORKERS = 0
EVAL_STEPS = 500
MAX_GRAD_NORM = 0.5
TOP_K_PASSAGES = 3

# RAG-specific configurations
RETRIEVER_MODEL = 'facebook/dpr-question_encoder-single-nq-base'
PASSAGE_MODEL = 'facebook/dpr-ctx_encoder-single-nq-base'

def clean_memory():
    """Clean up GPU memory."""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        torch.cuda.reset_peak_memory_stats()

def load_data(file_path, max_samples=None):
    """Load and preprocess the dataset with optional size limit."""
    logger.info(f"Loading data from {file_path}...")
    try:
        df = pd.read_csv(
            file_path,
            encoding='ISO-8859-1',
            on_bad_lines='skip',
            nrows=max_samples,
            low_memory=False
        )
        df.drop(['meeting_id', 'type', 'city'], axis=1, errors='ignore', inplace=True)

        df = df.apply(lambda x: x.str.strip() if isinstance(x, str) else x)
        df = df.fillna('')
        logger.info(f"Loaded {len(df)} rows after cleaning.")
        return df
    except Exception as e:
        logger.error(f"Error loading data: {str(e)}")
        raise

def calculate_rouge_scores(predicted_summary, reference_summary):
    """Calculate ROUGE scores for a pair of summaries."""
    try:
        scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
        scores = scorer.score(reference_summary, predicted_summary)
        return {
            'rouge1': scores['rouge1'].fmeasure,
            'rouge2': scores['rouge2'].fmeasure,
            'rougeL': scores['rougeL'].fmeasure
        }
    except Exception as e:
        logger.error(f"Error calculating ROUGE scores: {str(e)}")
        return {'rouge1': 0.0, 'rouge2': 0.0, 'rougeL': 0.0}

def create_word_cloud(text):
    """Generate and display word cloud from text."""
    try:
        doc = nlp(text)
        processed_text = ' '.join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])

        wordcloud = WordCloud(
            width=800,
            height=400,
            background_color='white',
            max_words=100,
            collocations=False
        ).generate(processed_text)

        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Generated Summaries')
        plt.show()
    except Exception as e:
        logger.error(f"Error creating word cloud: {str(e)}")

def print_summary_analysis(original_text, generated_summary, reference_summary=None):
    """Print detailed analysis of the summary."""
    try:
        print("\n" + "="*80)
        print("TEXT SUMMARIZATION ANALYSIS")
        print("="*80)

        print("\nORIGINAL TEXT:")
        print("-"*40)
        print(original_text[:500] + "..." if len(original_text) > 500 else original_text)

        print("\nGENERATED SUMMARY:")
        print("-"*40)
        print(generated_summary)

        if reference_summary:
            print("\nREFERENCE SUMMARY:")
            print("-"*40)
            print(reference_summary)

            print("\nROUGE SCORES:")
            print("-"*40)
            scores = calculate_rouge_scores(generated_summary, reference_summary)
            for metric, score in scores.items():
                print(f"{metric}: {score:.4f}")

        # Calculate and print statistics
        original_words = len(original_text.split())
        summary_words = len(generated_summary.split())
        compression_ratio = (summary_words / original_words) * 100 if original_words > 0 else 0

        print("\nSTATISTICS:")
        print("-"*40)
        print(f"Original text length: {original_words} words")
        print(f"Summary length: {summary_words} words")
        print(f"Compression ratio: {compression_ratio:.1f}%")
    except Exception as e:
        logger.error(f"Error in summary analysis: {str(e)}")


def plot_rouge_histogram(rouge_scores):
    """Plot a histogram of average ROUGE scores."""
    try:
        plt.figure(figsize=(8, 6))
        metrics = list(rouge_scores.keys())
        scores = list(rouge_scores.values())
        plt.bar(metrics, scores, color=['#2ecc71', '#3498db', '#9b59b6'])
        plt.title('Average ROUGE Scores')
        plt.xlabel('ROUGE Metric')
        plt.ylabel('Average Score')
        plt.ylim(0, 1)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
    except Exception as e:
        logger.error(f"Error plotting ROUGE histogram: {str(e)}")

class RAGRetriever:
    """RAG retrieval component for enhancing summarization with relevant context."""
    def __init__(self, passages: List[str]):
        self.passages = passages

        # Initialize DPR models
        self.question_encoder = DPRQuestionEncoder.from_pretrained(RETRIEVER_MODEL)
        self.context_encoder = DPRContextEncoder.from_pretrained(PASSAGE_MODEL)
        self.question_tokenizer = DPRQuestionEncoderTokenizer.from_pretrained(RETRIEVER_MODEL)
        self.context_tokenizer = DPRContextEncoderTokenizer.from_pretrained(PASSAGE_MODEL)

        if torch.cuda.is_available():
            self.question_encoder = self.question_encoder.cuda()
            self.context_encoder = self.context_encoder.cuda()

        # Create FAISS index for fast similarity search
        self._build_index()

    def _build_index(self):
        """Build FAISS index from passage embeddings."""
        logger.info("Building FAISS index for passages...")

        # Encode passages
        passage_embeddings = []
        for passage in tqdm(self.passages, desc="Encoding passages"):
            inputs = self.context_tokenizer(
                passage,
                max_length=MAX_LENGTH,
                padding=True,
                truncation=True,
                return_tensors='pt'
            )

            if torch.cuda.is_available():
                inputs = {k: v.cuda() for k, v in inputs.items()}

            with torch.no_grad():
                embedding = self.context_encoder(**inputs).pooler_output
                passage_embeddings.append(embedding.cpu().numpy())

        passage_embeddings = np.vstack(passage_embeddings)

        # Create FAISS index
        self.dimension = passage_embeddings.shape[1]
        self.index = faiss.IndexFlatIP(self.dimension)
        self.index.add(passage_embeddings)

        logger.info(f"FAISS index built with {len(self.passages)} passages")

    def retrieve(self, query: str, k: int = TOP_K_PASSAGES) -> List[str]:
        """Retrieve relevant passages for a query."""
        # Encode query
        inputs = self.question_tokenizer(
            query,
            max_length=MAX_LENGTH,
            padding=True,
            truncation=True,
            return_tensors='pt'
        )

        if torch.cuda.is_available():
            inputs = {k: v.cuda() for k, v in inputs.items()}

        with torch.no_grad():
            query_embedding = self.question_encoder(**inputs).pooler_output.cpu().numpy()

        # Search FAISS index
        scores, indices = self.index.search(query_embedding, k)

        # Return retrieved passages
        return [self.passages[idx] for idx in indices[0]]

class MemoryEfficientMeetingDataset(torch.utils.data.Dataset):
    """Memory-efficient dataset implementation with RAG support."""
    def __init__(self, df, tokenizer, max_length, retriever=None):
        self.df = df
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.retriever = retriever

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        text = str(self.df.iloc[idx, 0])[:self.max_length * 4]
        summary = str(self.df.iloc[idx, 1])[:self.max_length * 2]

        if self.retriever:
            # Retrieve relevant passages using the retriever
            retrieved_passages = self.retriever.retrieve(text)
            context = " | ".join(retrieved_passages)
            text = f"context: {context} text: {text}"

        input_encoding = self.tokenizer(
            f"summarize: {text}",
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        target_encoding = self.tokenizer(
            summary,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        return {
            'input_ids': input_encoding['input_ids'].squeeze(),
            'attention_mask': input_encoding['attention_mask'].squeeze(),
            'labels': target_encoding['input_ids'].squeeze()
        }


def train_model(max_train_samples=1000, max_val_samples=100):
    """Train the T5 model with memory optimizations and RAG."""
    try:
        clean_memory()

        train_df = load_data("/content/sample_data/train.csv", max_samples=max_train_samples)
        val_df = load_data("/content/sample_data/val.csv", max_samples=max_val_samples)

        # Initialize RAG retriever with training passages
        retriever = RAGRetriever(train_df.iloc[:, 0].tolist())

        tokenizer = T5Tokenizer.from_pretrained(MODEL)
        model = T5ForConditionalGeneration.from_pretrained(
            MODEL,
            torch_dtype=torch.bfloat16,
            low_cpu_mem_usage=True
        )

        if torch.cuda.is_available():
            model = model.cuda()

        train_dataset = MemoryEfficientMeetingDataset(train_df, tokenizer, MAX_LENGTH, retriever)
        val_dataset = MemoryEfficientMeetingDataset(val_df, tokenizer, MAX_LENGTH, retriever)

        clean_memory()

        training_args = TrainingArguments(
            output_dir=OUT_DIR,
            num_train_epochs=EPOCHS,
            per_device_train_batch_size=BATCH_SIZE,
            per_device_eval_batch_size=BATCH_SIZE,
            warmup_steps=100,
            weight_decay=0.01,
            logging_dir="./logs",
            logging_steps=50,
            eval_strategy='steps',
            eval_steps=EVAL_STEPS,
            save_strategy='steps',
            save_steps=EVAL_STEPS,
            save_total_limit=1,
            learning_rate=5e-5,
            dataloader_num_workers=NUM_WORKERS,
            bf16=True,
            gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
            load_best_model_at_end=True,
            metric_for_best_model="rouge1",
            greater_is_better=True,
            max_grad_norm=MAX_GRAD_NORM,
            report_to="none"
        )

        trainer = Trainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
        )

        logger.info("Starting model training...")
        trainer.train()

        return model, tokenizer, val_df, retriever

    except Exception as e:
        logger.error(f"Error during training: {str(e)}")
        raise

def generate_summary(text, model, tokenizer, retriever=None, max_length=64, prompt=None):
    """Generate a memory-efficient summary with RAG support."""
    try:
        if retriever:
            # Retrieve relevant passages using the retriever
            retrieved_passages = retriever.retrieve(text)
            context = " | ".join(retrieved_passages)
            text = f"context: {context} text: {text}"  # Combine retrieved context with the original text

        if prompt:
            text = f"{prompt} {text}"  # Add a prompt if provided

        # Tokenize the input text
        inputs = tokenizer.encode(
            text[:max_length * 4],  # Ensure the text is a string
            return_tensors='pt',
            max_length=max_length,
            truncation=True
        )

        # Move inputs to GPU if available
        if torch.cuda.is_available():
            inputs = inputs.cuda()

        # Generate the summary
        summary_ids = model.generate(
            inputs,
            max_length=max_length,
            num_beams=2,
            length_penalty=1.0,
            early_stopping=True,
            no_repeat_ngram_size=2
        )

        # Decode the generated tokens into text
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary
    except Exception as e:
        logger.error(f"Error generating summary: {str(e)}")
        return ""


def main():
    """Main execution pipeline with RAG-enhanced analysis and visualization."""
    try:
        logger.info("Starting training with RAG and memory optimizations...")
        model, tokenizer, val_data, retriever = train_model(max_train_samples=1000, max_val_samples=100)

        logger.info("Generating summaries and analyzing results...")
        val_summaries = []
        all_rouge_scores = []
        combined_summaries_text = ""

        # Summarization and analysis loop
        for i in tqdm(range(len(val_data)), desc="Summarizing and Analyzing"):
            # Get original text and reference summary
            original_text = val_data.iloc[i, 0]
            reference_summary = val_data.iloc[i, 1]

            # Generate summary using RAG
            generated_summary = generate_summary(original_text, model, tokenizer, retriever)

            # Process valid summaries
            if generated_summary:
                val_summaries.append(generated_summary)
                combined_summaries_text += " " + generated_summary

                # Calculate ROUGE scores
                scores = calculate_rouge_scores(generated_summary, reference_summary)
                all_rouge_scores.append(scores)

                # Print detailed analysis for the first 5 examples
                if i < 5:
                    print_summary_analysis(original_text, generated_summary, reference_summary)

                # Periodic memory cleanup
                if i % 10 == 0:
                    clean_memory()

        # Calculate and display average ROUGE scores
        if all_rouge_scores:
            avg_scores = {
                metric: np.mean([score[metric] for score in all_rouge_scores])
                for metric in ['rouge1', 'rouge2', 'rougeL']
            }

            print("\n" + "="*80)
            print("OVERALL EVALUATION METRICS")
            print("="*80)
            print("\nAverage ROUGE Scores:")
            for metric, score in avg_scores.items():
                print(f"{metric}: {score:.4f}")

            # Plot histogram of ROUGE scores
            logger.info("Generating ROUGE score histogram...")
            plot_rouge_histogram(avg_scores)
            print("\nHistogram of average ROUGE scores displayed.")

        # Generate and display word cloud
        if combined_summaries_text:
            logger.info("Generating word cloud visualization...")
            create_word_cloud(combined_summaries_text)
            print("\nWord cloud has been displayed.")

        logger.info("Process completed successfully.")

    except Exception as e:
        logger.error(f"An error occurred: {str(e)}")
        raise
    finally:
        clean_memory()


if __name__ == "__main__":
    main()
```

#### Using t5-small Model 

### Details
The T5-small model is a variant of the Text-to-Text Transfer Transformer (T5), which was introduced by Google.T5 (Text-to-Text Transfer Transformer) treats every NLP task as a text-to-text problem. This unifies the approach to various tasks like classification, summarization, and translation under a single framework.
Key Features of T5-Small:

Compact Size: It's a smaller version of the full T5 model, making it more suitable for deployment on devices with limited resources.   
Versatility: It can be adapted to a wide range of NLP tasks by fine-tuning on specific datasets.   
Efficiency: It's relatively efficient to train and use compared to larger language models.


```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

import torch
import numpy as np
import pandas as pd
import evaluate
from tqdm import tqdm
import logging
import gc
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback
)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from rouge_score import rouge_scorer
import spacy
from typing import List, Dict
from wordcloud import STOPWORDS

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Configuration
MODEL = 't5-small'
OUT_DIR = "./model_checkpoints"
BATCH_SIZE = 1
EPOCHS = 2
MAX_LENGTH = 64
GRADIENT_ACCUMULATION_STEPS = 32
NUM_WORKERS = 0
EVAL_STEPS = 500
MAX_GRAD_NORM = 0.5

def clean_memory():
    """Clean up GPU memory."""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        torch.cuda.reset_peak_memory_stats()

def load_data(file_path, max_samples=None):
    """Load and preprocess the dataset with optional size limit."""
    logger.info(f"Loading data from {file_path}...")
    try:
        df = pd.read_csv(
            file_path,
            encoding='ISO-8859-1',
            on_bad_lines='skip',
            nrows=max_samples,
            low_memory=False
        )
        df.drop(['meeting_id', 'type', 'city'], axis=1, errors='ignore', inplace=True)

        df = df.apply(lambda x: x.str.strip() if isinstance(x, str) else x)
        df = df.fillna('')
        logger.info(f"Loaded {len(df)} rows after cleaning.")
        return df
    except Exception as e:
        logger.error(f"Error loading data: {str(e)}")
        raise

def calculate_rouge_scores(predicted_summary, reference_summary):
    """Calculate ROUGE scores for a pair of summaries."""
    try:
        scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
        scores = scorer.score(reference_summary, predicted_summary)
        return {
            'rouge1': scores['rouge1'].fmeasure,
            'rouge2': scores['rouge2'].fmeasure,
            'rougeL': scores['rougeL'].fmeasure
        }
    except Exception as e:
        logger.error(f"Error calculating ROUGE scores: {str(e)}")
        return {'rouge1': 0.0, 'rouge2': 0.0, 'rougeL': 0.0}

def create_word_cloud(text):
    """Generate and display word cloud from text."""
    try:
        doc = nlp(text)
        processed_text = ' '.join([token.lemma_ for token in doc if not token.is_stop and not token.is_punct])

        wordcloud = WordCloud(
            width=800,
            height=400,
            background_color='white',
            max_words=100,
            collocations=False
        ).generate(processed_text)

        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Word Cloud of Generated Summaries')
        plt.show()

        # Generate ASCII art for word cloud
        print("\nASCII Word Cloud:\n")
        for word, freq in sorted(wordcloud.words_.items(), key=lambda x: -x[1])[:20]:
            print(f"{word:<20} {'#' * int(freq * 40)}")
    except Exception as e:
        logger.error(f"Error creating word cloud: {str(e)}")

def create_ascii_wordcloud(texts, width=80, height=20):
    """Generate and display ASCII word cloud in console."""
    import spacy
    from collections import Counter

    nlp = spacy.load('en_core_web_sm')

    # Process all texts
    words = []
    for text in texts:
        doc = nlp(text)
        words.extend([
            token.lemma_.lower() for token in doc
            if not token.is_stop and
            not token.is_punct and
            token.text.strip() and
            len(token.text) > 2
        ])

    # Calculate frequencies
    word_freq = Counter(words)
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
    max_freq = max(freq for _, freq in sorted_words)

    # Display results
    print("\nWord Frequency Analysis")
    print("=" * width)
    print(f"{'Word':<20} {'Frequency':<10} {'Distribution':<50}")
    print("-" * width)

    # Show top words with bar visualization
    for word, freq in sorted_words[:20]:
        scaled_freq = int((freq / max_freq) * 40)
        bar = "█" * scaled_freq
        percentage = (freq / sum(freq for _, freq in sorted_words)) * 100
        print(f"{word:<20} {freq:<10} {bar:<40} {percentage:>6.2f}%")

    print("=" * width)
    print(f"Total unique words: {len(word_freq)}")
    print(f"Total words processed: {sum(word_freq.values())}")

    return word_freq

def print_summary_analysis(original_text, generated_summary, reference_summary=None):
    """Print detailed analysis of the summary."""
    try:
        print("\n" + "="*80)
        print("TEXT SUMMARIZATION ANALYSIS")
        print("="*80)

        print("\nORIGINAL TEXT:")
        print("-"*40)
        print(original_text[:500] + "..." if len(original_text) > 500 else original_text)

        print("\nGENERATED SUMMARY:")
        print("-"*40)
        print(generated_summary)

        if reference_summary:
            print("\nREFERENCE SUMMARY:")
            print("-"*40)
            print(reference_summary)

            print("\nROUGE SCORES:")
            print("-"*40)
            scores = calculate_rouge_scores(generated_summary, reference_summary)
            for metric, score in scores.items():
                print(f"{metric}: {score:.4f}")

        # Calculate and print statistics
        original_words = len(original_text.split())
        summary_words = len(generated_summary.split())
        compression_ratio = (summary_words / original_words) * 100 if original_words > 0 else 0

        print("\nSTATISTICS:")
        print("-"*40)
        print(f"Original text length: {original_words} words")
        print(f"Summary length: {summary_words} words")
        print(f"Compression ratio: {compression_ratio:.1f}%")
    except Exception as e:
        logger.error(f"Error in summary analysis: {str(e)}")


def plot_rouge_histogram(rouge_scores):
    """Plot a histogram of average ROUGE scores."""
    try:
        plt.figure(figsize=(8, 6))
        metrics = list(rouge_scores.keys())
        scores = list(rouge_scores.values())
        plt.bar(metrics, scores, color=['#2ecc71', '#3498db', '#9b59b6'])
        plt.title('Average ROUGE Scores')
        plt.xlabel('ROUGE Metric')
        plt.ylabel('Average Score')
        plt.ylim(0, 1)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
    except Exception as e:
        logger.error(f"Error plotting ROUGE histogram: {str(e)}")

class MemoryEfficientMeetingDataset(torch.utils.data.Dataset):
    """Memory-efficient dataset implementation."""
    def __init__(self, df, tokenizer, max_length):
        self.df = df
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        text = str(self.df.iloc[idx, 0])[:self.max_length * 4]
        summary = str(self.df.iloc[idx, 1])[:self.max_length * 2]

        input_encoding = self.tokenizer(
            f"summarize: {text}",
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        target_encoding = self.tokenizer(
            summary,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        return {
            'input_ids': input_encoding['input_ids'].squeeze(),
            'attention_mask': input_encoding['attention_mask'].squeeze(),
            'labels': target_encoding['input_ids'].squeeze()
        }


def train_model(max_train_samples=1000, max_val_samples=100):
    """Train the T5 model."""
    try:
        clean_memory()

        train_df = load_data("/content/sample_data/train.csv", max_samples=max_train_samples)
        val_df = load_data("/content/sample_data/val.csv", max_samples=max_val_samples)

        tokenizer = T5Tokenizer.from_pretrained(MODEL)
        model = T5ForConditionalGeneration.from_pretrained(
            MODEL,
            torch_dtype=torch.bfloat16,
            low_cpu_mem_usage=True
        )

        if torch.cuda.is_available():
            model = model.cuda()

        train_dataset = MemoryEfficientMeetingDataset(train_df, tokenizer, MAX_LENGTH)
        val_dataset = MemoryEfficientMeetingDataset(val_df, tokenizer, MAX_LENGTH)

        clean_memory()

        training_args = TrainingArguments(
            output_dir=OUT_DIR,
            num_train_epochs=EPOCHS,
            per_device_train_batch_size=BATCH_SIZE,
            per_device_eval_batch_size=BATCH_SIZE,
            warmup_steps=100,
            weight_decay=0.01,
            logging_dir="./logs",
            logging_steps=50,
            eval_strategy='steps',
            eval_steps=EVAL_STEPS,
            save_strategy='steps',
            save_steps=EVAL_STEPS,
            save_total_limit=1,
            learning_rate=5e-5,
            dataloader_num_workers=NUM_WORKERS,
            bf16=True,
            gradient_accumulation_steps=GRADIENT_ACCUMULATION_STEPS,
            load_best_model_at_end=True,
            metric_for_best_model="rouge1",
            greater_is_better=True,
            max_grad_norm=MAX_GRAD_NORM,
            report_to="none"
        )

        trainer = Trainer(
            model=model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
        )

        logger.info("Starting model training...")
        trainer.train()

        return model, tokenizer, val_df

    except Exception as e:
        logger.error(f"Error during training: {str(e)}")
        raise

def generate_summary(text, model, tokenizer, max_length=64, prompt=None):
    """Generate a memory-efficient summary."""
    try:
        if prompt:
            text = f"{prompt} {text}"  # Add a prompt if provided

        # Tokenize the input text
        inputs = tokenizer.encode(
            text[:max_length * 4],  # Ensure the text is a string
            return_tensors='pt',
            max_length=max_length,
            truncation=True
        )

        # Move inputs to GPU if available
        if torch.cuda.is_available():
            inputs = inputs.cuda()

        # Generate the summary
        summary_ids = model.generate(
            inputs,
            max_length=max_length,
            num_beams=2,
            length_penalty=1.0,
            early_stopping=True,
            no_repeat_ngram_size=2
        )

        # Decode the generated tokens into text
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        return summary
    except Exception as e:
        logger.error(f"Error generating summary: {str(e)}")
        return ""


def main():
    """Main execution pipeline."""
    try:
        logger.info("Starting training and analysis...")
        model, tokenizer, val_data = train_model(max_train_samples=1000, max_val_samples=100)

        logger.info("Generating summaries and analyzing results...")
        val_summaries = []
        all_rouge_scores = []
        combined_summaries_text = []

        # Summarization and analysis loop
        for i in tqdm(range(len(val_data)), desc="Summarizing and Analyzing"):
            # Get original text and reference summary
            original_text = val_data.iloc[i, 0]
            reference_summary = val_data.iloc[i, 1]

            # Generate summary
            generated_summary = generate_summary(original_text, model, tokenizer)

            # Process valid summaries
            if generated_summary:
                val_summaries.append(generated_summary)
                combined_summaries_text.append(generated_summary)

                # Calculate ROUGE scores
                scores = calculate_rouge_scores(generated_summary, reference_summary)
                all_rouge_scores.append(scores)

                # Print detailed analysis for the first 5 examples
                if i < 5:
                    print_summary_analysis(original_text, generated_summary, reference_summary)

                # Periodic memory cleanup
                if i % 10 == 0:
                    clean_memory()

        # Calculate and display average ROUGE scores
        if all_rouge_scores:
            avg_scores = {
                metric: np.mean([score[metric] for score in all_rouge_scores])
                for metric in ['rouge1', 'rouge2', 'rougeL']
            }

            print("\n" + "="*80)
            print("OVERALL EVALUATION METRICS")
            print("="*80)
            print("\nAverage ROUGE Scores:")
            for metric, score in avg_scores.items():
                print(f"{metric}: {score:.4f}")

            # Plot histogram of ROUGE scores
            logger.info("Generating ROUGE score histogram...")
            plot_rouge_histogram(avg_scores)
            print("\nHistogram of average ROUGE scores displayed.")

        # Generate and display ASCII word cloud
        if combined_summaries_text:
            logger.info("Generating ASCII word cloud visualization...")
            create_ascii_wordcloud(combined_summaries_text)
            print("\nASCII word cloud has been displayed.")

        logger.info("Process completed successfully.")

    except Exception as e:
        logger.error(f"An error occurred: {str(e)}")
        raise
    finally:
        clean_memory()


if __name__ == "__main__":
    main()
```
